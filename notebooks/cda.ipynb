```python
import matplotlib.pyplot as plt
import json
from transform.gdc.gdc_transformer import make_reader
from attrdict import AttrDict
from transform.pdc.pdc_harvester import load
from transform.tcia.tcia_harvester import load  as tcia_loader
from matplotlib_venn import venn3, venn2

gdc_cases = [ AttrDict(json.loads(c)) for c in make_reader('Case.Vertex')]
pdc_cases = load('allCases')
tcia_cases = [o for o in tcia_loader('patients')]
tcia_patient_studies = [o for o in tcia_loader('patient_study')]

gdc_primary_diagnoses = set()
for c in gdc_cases:
    if not c.data.gdc_attributes.diagnoses:
        continue
    for d in c.data.gdc_attributes.diagnoses:
        gdc_primary_diagnoses.add(d.primary_diagnosis)

pdc_primary_diagnoses = set()
for c in pdc_cases:
    if not c.diagnoses:
        continue
    for d in c.diagnoses:
        pdc_primary_diagnoses.add(d.primary_diagnosis)        
        

tcia_primary_diagnoses = set([])        
# Note: no diagnoses in tcia
# tcia_cases[0]
# AttrDict({'PatientID': 'TCGA-08-0244', 'PatientName': '', 'PatientSex': 'M', 'Collection': 'TCGA-GBM'})
# tcia_patient_studies[0]
# AttrDict({'Collection': 'TCGA-GBM', 'PatientID': 'TCGA-06-0119', 'PatientName': 'TCGA-06-0119', 'PatientSex': 'F', 'StudyInstanceUID': '1.3.6.1.4.1.14519.5.2.1.4591.4001.283171458436570880779969690543', 'StudyDate': '2003-12-26', 'SeriesCount': 16})

#venn3([gdc_primary_diagnoses, pdc_primary_diagnoses, tcia_primary_diagnoses], set_labels = ('gdc',  'pdc', 'tcia'))

venn2([gdc_primary_diagnoses, pdc_primary_diagnoses], set_labels = ('gdc',  'pdc'))
plt.savefig('notebooks/figures/primary_diagnoses.png')
plt.show()
```
